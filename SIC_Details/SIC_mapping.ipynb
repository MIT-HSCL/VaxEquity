# Counties

```python
from dbfread import DBF
import pandas as pd

# Open the DBF file
table = DBF('SIC_County_Details.dbf')

# Create a list of dictionaries for each record
records = []
for record in table:
    records.append(dict(record))

# Convert the list of dictionaries to a pandas DataFrame
df = pd.DataFrame(records)
```

```python
SIC_counties = df.pivot(index='Join_ID', columns='SIC').rename(sic_mapping, axis=1).fillna(0).astype(int).droplevel(0, axis=1)
SIC_counties
```

```python
to_remove_prefixed = ["x_" + c  for c in to_remove]
SIC_counties.rename(columns = dict(zip(to_remove, to_remove_prefixed)), inplace=True)
```

```python
#pd.DataFrame(SIC_counties.sum()).to_excel("SIC_Export.xlsx")
```

```python
#SIC_counties[to_keep].sum(axis=1)
```

```python
SIC_counties["Point Coun"] = SIC_counties[to_keep].sum(axis=1)
```

```python
SIC_counties
```

```python
SIC_counties = SIC_counties.reset_index()[["Join_ID", "Point Coun"]].copy()
```

```python
SIC_counties["Point Coun"].isna().sum()
```

```python
facilities = gpd.read_file("../Counties_Medical_Facilities/Counties_Medical_Facilities_Detailed.shp", include_fields=["FIPS", "Join_ID"])
```

```python
facilities = facilities[["FIPS", "Join_ID"]].copy()
```

```python
facilities = facilities.set_index('Join_ID').join(SIC_counties.set_index('Join_ID'))
```

```python
facilities.to_csv("../Data_Preparation/medical_facilities_counties.csv")
```
